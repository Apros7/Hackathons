```python
import pandas as pd
import matplotlib.pyplot as plt
from ipynb import display
```

```python
all_data = pd.read_excel("../Case/Health Data from International Diabetes Foundation.xlsx", header=1, sheet_name=None)
```

```python
all_data.keys()
```

```python
def get_value(v):
    if v == "-":
        return 0
    if isinstance(v, str):
       v = v.replace(",", "") 
    return float(v)

def get_top_bad_countries(key, limit, reverse = False):
    data = all_data[key]
    data["2021int"] = data[2021].apply(get_value)
    no_regions = data[data.Region != data["Country/Territory"]]
    no_regions = no_regions.sort_values(by = "2021int", ascending=reverse)
    display(no_regions.head(50))
    display(no_regions[no_regions["2021int"] < limit]["Region"].value_counts() if reverse else no_regions[no_regions["2021int"] > limit]["Region"].value_counts())
```

```python
get_top_bad_countries("Diabetes prevalence (%)", 15.0)
```

```python
get_top_bad_countries("People with diabetes (1000s)", 3000)
```

```python
get_top_bad_countries("Diabetes-related deaths (propo)", 10)
```

```python
get_top_bad_countries("Diabetes costs per person (USD)", 10, reverse = True)
```
